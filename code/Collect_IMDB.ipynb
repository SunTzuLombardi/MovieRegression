```python
import pandas as pd
import imdb_api as ia
import time
import json
from json import JSONEncoder
from imdb_api import Movie
from io import StringIO
```

```python
moviesjsonfile = './data/imdb_movies.json'
```

```python
nb_df = pd.read_csv('./data/final_merged_tidy.csv', encoding='utf-8', index_col=False)
```

```python
# A specialised JSONEncoder that encodes Movie
# objects as JSON
class MovieEncoder(JSONEncoder):

    def default(self, object):

        if isinstance(object, Movie):
            return object.__dict__
        else:
            # call base class implementation which takes care of
            # raising exceptions for unsupported types
            return json.JSONEncoder.default(self, object)
```

```python
#RUNS in 17 MINS for 3000 movies
#collect IMDB movie objects isung imdb_api
def collect_imdb_movies():
    counter = 0
    movies = []
    for id in nb_df.tconst:

        print(counter,id)
        movies.append(ia.search_by_id(id))
        counter +=1
        if counter%10 == 0:
            time.sleep(1)
    return movies
```

```python
#list of MOVIE objects to json file
def movies_to_jsonfile(movies):
    movie_df = pd.DataFrame()
    for movie in movies:
        ## Use the MovieEncoder to decode a Movie object
        jsn = pd.read_json(StringIO(MovieEncoder().encode(movie)),
                           typ='series', encoding='utf-8')
        fr = jsn.to_frame()
        movie_df = movie_df.append(fr.T, ignore_index=True)
    movie_df.to_json(moviesjsonfile)
```

```python
def load_jsonfile_todf():
    with open(moviesjsonfile) as f:
        #load json to dict to createDF
        json_df = pd.DataFrame(json.load(f))
    return json_df
```

```python
def create_actor_influence_df():
    actor_df = pd.DataFrame(columns=['tconst','title','released_date','actor','dombox','adj_dombox'])
    for row in shortmerged_df.iterrows():
        tconst = row[1][0]
        title = row[1][1]
        released_date = row[1][2]
        actors = row[1][3].split(',')
        dombox = row[1][4]
        adj_dombox = row[1][5]
        for actor in actors:
            actor_df = actor_df.append({
            'tconst': tconst,
            'title':  title,
            'released_date' : released_date,
            'actor': actor,
            'dombox': dombox, 
            'adj_dombox':adj_dombox
          }, ignore_index=True)     
    return actor_df
```

```python
actor_df = create_actor_influence_df()
```

```python
actor_df.sort_values(by=['actor', 'released_date'],ascending=[True,True],inplace=True)
```

```python
actor_df.released_date = actor_df.released_date.replace(to_replace='N/A',value='',)
```
